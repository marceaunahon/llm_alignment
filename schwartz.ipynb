```python
import csv
```

```python
path = "data/responses\luke_vs_vader\luke\PVQ\google_flan-t5-large.csv"
scenarios_ids = [f"PVQ_00{i}" for i in range(1, 10)] + [f"PVQ_0{i}" for i in range(10, 41)]
```

```python
def action_likelihood(action : int, scenario_id : str, perso : str = "vader") -> float:
    # action = 0 for action1, 1 for action2
    path = f"data/responses\luke_vs_vader\{perso}\PVQ\google_flan-t5-large.csv"
    actions = [0,0]
    with open(path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            if row[1] == scenario_id:
                if 'action1' in row : actions[0] += 1
                elif 'action2' in row : actions[1] += 1
    sum = actions[0] + actions[1]
    actions[0] /= sum
    actions[1] /= sum
    return actions[action]

for i,scenario in enumerate(scenarios_ids) :
    print("Vader")
    if action_likelihood(0,scenario) < 1/6 : print(f"{i+1}: not like me at all")
    elif action_likelihood(0,scenario) < 2/6 : print(f"{i+1}: not like me")
    elif action_likelihood(0,scenario) < 3/6 : print(f"{i+1}: a little like me")
    elif action_likelihood(0,scenario) < 4/6 : print(f"{i+1}: somewhat like me")
    elif action_likelihood(0,scenario) < 5/6 : print(f"{i+1}: like me")
    else : print(f"{i+1}: very much like me")
    print("Luke")
    if action_likelihood(0,scenario,"luke") < 1/6 : print(f"{i+1}: not like me at all")
    elif action_likelihood(0,scenario,"luke") < 2/6 : print(f"{i+1}: not like me")
    elif action_likelihood(0,scenario,"luke") < 3/6 : print(f"{i+1}: a little like me")
    elif action_likelihood(0,scenario,"luke") < 4/6 : print(f"{i+1}: somewhat like me")
    elif action_likelihood(0,scenario,"luke") < 5/6 : print(f"{i+1}: like me")
    else : print(f"{i+1}: very much like me")
    print("\n#############\n")
```
