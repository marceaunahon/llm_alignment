```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def gather_column_results(csv_path : str, column_index : int = 14) -> list[any]:
    results = []
    with open(csv_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            results.append(row[column_index])
    actions_dict = {'action1': results.count('action1')/len(results), 'action2': results.count('action2')/len(results)}
    return actions_dict

def plot_bar(dict: dict[str:int], x_labels : list[str], scenario : str) -> None:

    ind = np.arange(len(x_labels))
    plt.bar(dict.keys(), dict.values(), width=0.6,align='center')
    plt.xticks(ind, x_labels)
    plt.title(scenario)
    plt.xlabel("Actions")
    plt.ylabel("Percentage of actions taken")
    plt.show()
    print(dict)
```

```python
gandhi_path = 'data/responses/gandhi/gandhi/google_flan-t5-small.csv'
beggar_path = 'data/responses/beggar/beggar/google_flan-t5-small.csv'
beggar_near_path = 'data/responses/beggar_near/beggar_near/google_flan-t5-small.csv'
beggar_near_intentionally_path = 'data/responses/beggar_near_intentionally/beggar_near_intentionally/google_flan-t5-small.csv'
no_path = 'data/responses/no/no/google_flan-t5-small.csv'
actions1 = gather_column_results(gandhi_path, 14)
actions2 = gather_column_results(beggar_path, 14)
actions3 = gather_column_results(beggar_near_path, 14)
actions4 = gather_column_results(beggar_near_intentionally_path, 14)
actions5 = gather_column_results(no_path, 14)
plot_bar(actions1, ["violated", "did not violate"], "Gandhi")
plot_bar(actions2, ["violated", "did not violate"], "Beggar")
plot_bar(actions3, ["violated", "did not violate"], "Beggar near")
plot_bar(actions4, ["violated", "did not violate"], "Beggar near intentionally")
plot_bar(actions5, ["violated", "did not violate"], "No")
```
