```python
embedding_dict = {}

import csv
import pandas as pd
import numpy as np

# read in the data
df = pd.read_csv('english Dictionary.csv')
words = df['word'].tolist()
print(len(words))

words2 = list(set(words))
print(len(words2))


with open('gpt_embeddings.csv', 'a', newline='', encoding='utf-8') as csvfile:
    # Créer un objet writer
    writer = csv.writer(csvfile)

    # Boucler à travers les mots
    for word in words2:
        if not isinstance(word, str):
            print(word)
            continue

        # Obtenir l'incorporation du mot
        embedding = client.embeddings.create(input=word, model="text-embedding-3-small").data[0].embedding

        # Écrire la ligne dans le fichier CSV
        writer.writerow([word, embedding])

        # Ajouter l'embedding au dictionnaire
        embedding_dict[word] = embedding

        if words2.index(word) % 10000 == 0:
            print(words2.index(word))

```
