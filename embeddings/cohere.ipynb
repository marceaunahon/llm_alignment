```python
def cosine_similarity(word1 : str, word2 : str) -> float:
  response = co.embed(
    texts=[word1, word2],
    model='embed-english-v3.0',
    input_type='classification'
  )
  a = response.embeddings[0]
  b = response.embeddings[1]
  return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

```
